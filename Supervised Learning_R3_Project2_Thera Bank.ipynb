```python
thera_train = thera_df.drop(["ID","ZIP Code","Personal Loan"],axis=1)
thera_train_standard =(thera_train-thera_train.mean())/thera_train.std()
Y = thera_df["Personal Loan"]
from sklearn.model_selection import train_test_split
from sklearn import metrics
#xtrain,xtest,ytrain,ytest=train_test_split(thera_train,Y,test_size=0.3,random_state=0)
xtrain,xtest,ytrain,ytest=train_test_split(thera_train_standard,Y,test_size=0.3,random_state=0)
Y.value_counts()
```

```python
from sklearn.linear_model import LogisticRegression
model_LR = LogisticRegression()
model_LR.fit(xtrain,ytrain)
ypred1=model_LR.predict(xtest)
acc1=metrics.accuracy_score(ytest,ypred1)
print(acc1)
cm1=metrics.confusion_matrix(ytest,ypred1)
print(cm1)
print(metrics.classification_report(ytest, ypred1))
```

```python
from sklearn.naive_bayes import GaussianNB
model_NB = GaussianNB()
model_NB.fit(xtrain, ytrain)
predict_NB = model_NB.predict(xtest)
acc1=metrics.accuracy_score(ytest,predict_NB)
print(acc1)
cm1=metrics.confusion_matrix(ytest,predict_NB)
print(cm1)
print(metrics.classification_report(ytest, predict_NB))
```

```python
from sklearn.neighbors import KNeighborsClassifier  
model_KNN = KNeighborsClassifier(n_neighbors=6)  
model_KNN.fit(xtrain, ytrain) 
predict_KNN = model_KNN.predict(xtest)
acc1=metrics.accuracy_score(ytest,predict_KNN)
print(acc1)
cm1=metrics.confusion_matrix(ytest,predict_KNN)
print(cm1)
print(metrics.classification_report(ytest, predict_KNN))
```
