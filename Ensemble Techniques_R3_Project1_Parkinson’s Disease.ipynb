```python
p_train = p_df.drop(["name","status"],axis=1)
```

```python
# Due to high deviation in the values. We are first normalizing the dataset
nor_df =(p_train-p_train.mean())/p_train.std()
```

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
Y=p_df['status']
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(p_train,Y,test_size=0.3,random_state=0)
```

```python
#Create the model using â€œentropyâ€ method of reducing the entropy and fit it to training data
from sklearn.tree import DecisionTreeClassifier
DT_model=DecisionTreeClassifier(criterion = 'entropy',max_depth=None)
DT_model.fit(xtrain,ytrain)
```

```python
from sklearn import metrics
ypred=DT_model.predict(xtest)
acc=metrics.accuracy_score(ytest,ypred)
print(acc)
cm=metrics.confusion_matrix(ytest,ypred)
print(cm)
```

```python
from sklearn.ensemble import RandomForestClassifier
RF_model = RandomForestClassifier(n_estimators=51,criterion='entropy',max_depth=None,random_state=0 )
RF_model.fit(xtrain, ytrain)
ypred_RF = RF_model.predict(xtest)
acc_rf=metrics.accuracy_score(ytest,ypred_RF)
print(acc_rf)
cm_rf=metrics.confusion_matrix(ytest,ypred_RF)
print(cm_rf)
```

```python
#Use regularization parameters of max_depth, min_sample_leaf to recreate the model.
#What is the impact on the model accuracy? How does regularization help? (20 points) 
DT_model=DecisionTreeClassifier(criterion = 'entropy',max_depth=5)
DT_model.fit(xtrain,ytrain)
ypred=DT_model.predict(xtest)
acc=metrics.accuracy_score(ytest,ypred)
print(acc)
cm=metrics.confusion_matrix(ytest,ypred)
print(cm)
```

```python
from sklearn.ensemble import RandomForestClassifier
RF_model = RandomForestClassifier(n_estimators=52,criterion='entropy',max_depth=5,random_state=0 )
RF_model.fit(xtrain, ytrain)
ypred_RF = RF_model.predict(xtest)
acc_rf=metrics.accuracy_score(ytest,ypred_RF)
print(acc_rf)
cm_rf=metrics.confusion_matrix(ytest,ypred_RF)
print(cm_rf)
```

```python
from sklearn.ensemble import ExtraTreesClassifier
forest = ExtraTreesClassifier(n_estimators=250,random_state=0)
forest.fit(xtrain,ytrain)
importances=forest.feature_importances_
```

```python
import numpy as np
std = np.std([tree.feature_importances_ for tree in forest.estimators_],
             axis=0)
indices = np.argsort(importances)[::-1]
print("Feature ranking:")
for f in range(xtest.shape[1]):
    print("%d. feature %d (%f)" % (f + 1, indices[f], importances[indices[f]]))
# Plot the feature importances of the forest
plt.figure()
plt.title("Feature importances")
plt.bar(range(xtrain.shape[1]), importances[indices],
       color="r", yerr=std[indices], align="center")
plt.xticks(range(xtrain.shape[1]), indices)
plt.xlim([-1, xtrain.shape[1]])
plt.show()
```
