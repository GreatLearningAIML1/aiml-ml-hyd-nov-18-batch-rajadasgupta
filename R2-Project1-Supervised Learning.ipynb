```python
import pandas as pd
airline_df = pd.read_csv('Airlinedata.csv')
```

```python
airline_df.describe()
```

```python
#Dropping all categorical variable and considering only continious variables to find outliers
from scipy import stats
import numpy as np
dist_z = np.abs(stats.zscore(airline_df[['Avgfare','Distance','AWP','MLA_MS','MLA_AF','LPA_MS','LPA_AF']]))
#As i am considering only absolute value hence not taking into consideration -3 to find outliers
print (np.where(dist_z > 3)[0])
airlinedf1  = airline_df.drop(airline_df.index[np.where(dist_z > 3)[0]])
```

```python
airlinedf1
# 34 outliers found which were dropped from the dataframe before running the model
```

```python
airlinedf1.corr()
```

```python
#Dropping all independent varibale with correalation less than 0.1 i.e. Average weekly passengers,
# Market leading airlines market share, Low price airlines market share
airlinedf1_X = airlinedf1[['Distance','MLA_AF','LPA_AF']]
airlinedf1_Y = airlinedf1[['Avgfare']].values.reshape(-1,1)
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
airlinedf1.plot.scatter(x='Distance', y='Avgfare');
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
airlinedf1.plot.scatter(x='MLA_AF', y='Avgfare');
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
airlinedf1.plot.scatter(x='LPA_AF', y='Avgfare');
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(airlinedf1_X, airlinedf1_Y, test_size=0.2, random_state = 1)
```

```python
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
linreg.fit(x_train, y_train)
print('Coeffcient is :', linreg.coef_)
```

```python
test_acc = linreg.score(x_test, y_test)
train_acc = linreg.score(x_train, y_train)
print ('Accuracy value for test dataset is', test_acc)
print ('Accuracy value for train dataset is', train_acc)
```
