### Find out the no of male & female employees in the company

*Hint: Use can use a condition to find out the no of 1s and 0s in gender and sum it*

```python
#Considering Male as 1 and Female as 2

malecnt = 0
femalecnt = 0
for num in Hrfile.iloc[:,12]:
    if num == 1:
        malecnt += 1
    else:
        femalecnt += 1
print ("Number of male:", malecnt)
print ("Number of female:", femalecnt)
```
