```python
#Reading the 30 samples into a dataframe
import pandas as pd
import numpy as np
salesdf = pd.read_csv('Sales.csv')
```

```python
#Printing the mean of two schemes
print ('The mean value of old schedme', salesdf['OS'].mean())
print ('The mean value of new schedme', salesdf['NS'].mean())
```

```python
from scipy.stats import ttest_1samp, ttest_ind, mannwhitneyu, levene, shapiro
from statsmodels.stats.power import ttest_power

#Determine the linearlity of the data
shapiro(salesdf['OS'])
```

```python
shapiro(salesdf['NS'])
```

```python
levene(salesdf['OS'], salesdf['NS'])
```

```python
t_statistic, p_value = ttest_ind(salesdf['NS'], salesdf['OS'])
print(t_statistic, p_value)
```

```python
pre = salesdf['OS']
post = salesdf['NS']
# paired t-test: doing two measurments on the same experimental unit
# e.g., before and after a treatment
t_statistic, p_value = ttest_1samp(post - pre, 0)
print(t_statistic, p_value)
```

```python
# Null Hypothesis: The mean of old scheme is equal to mean of new scheme.
# Since the P Value is higher than 0.05 hence we fail to reject the null hypothesis. This means that the change in
# scheme "Has NOT" improved or rasied the output.
```

```python
#Power of test
(np.mean(post) - np.mean(pre)) / np.sqrt(((30-1)*np.var(pre)+(30-1)*np.var(post)) / 30+30-2)
```

```python
#What is the p- value of the hypothesis test if we test for a difference of $5000?
t_statistic, p_value = ttest_1samp(post - pre, 4500)
print(t_statistic, p_value)
```
